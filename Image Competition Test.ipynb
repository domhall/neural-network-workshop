```python
%matplotlib inline
# %matplotlib ipympl
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=8,8

import numpy as np
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
```

```python
from keras.datasets import cifar10
import keras
from keras.models import Sequential, load_model
from keras.layers.core import Dense, Activation, Dropout, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D

import tensorflow as tf
```

```python
model = keras.models.load_model('competition_model')
```

```python
(_, _), (X_test, y_test) = cifar10.load_data()
```

```python
y_test = np_utils.to_categorical(y_test)
X_test_norm = X_test.astype('float32') / 255
```

```python
print("Test Accuracy:", model.evaluate(X_test_norm, y_test)[1])
```

```python
labels = {
    0: 'airplane',
    1: 'automobile',
    2: 'bird',
    3: 'cat',
    4: 'deer',
    5: 'dog',
    6: 'frog',
    7: 'horse',
    8: 'ship',
    9: 'truck'
}
```

```python
y_pred = model.predict(X_test_norm)
```

```python
fig = plt.figure()
j = 0
i = 0
while i < 9:
    if np.argmax(y_pred[i + j]) != np.argmax(y_test[i + j]):
        plt.subplot(3,3,i+1)
        plt.tight_layout()
        plt.imshow(X_test[i + j], cmap='hot', interpolation='none')
        plt.title("Pred: {} True: {}".format(labels[np.argmax(y_pred[i + j])], labels[np.argmax(y_test[i + j])]))
        plt.xticks([])
        plt.yticks([])
        i += 1
    else :
        j += 1
            
```
